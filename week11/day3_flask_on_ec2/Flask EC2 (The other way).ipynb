## Getting Flask Up and Running

#### The Rough and Ready version

ideally we would not run flask apps this way in production environments, but it will be perfect for just getting something up and running remotely


start an EC2 instance as previously - use the Anaconda3 Ubuntu community AMI



Make sure you open up port 80 to http traffic:



## Connect to your instance

`sudo apt-get update`  
`sudo apt-get install alien`  
`sudo apt install python-pip3`  
`sudo pip3 install -U pip numpy flask sklearn pandas`

once you have done this, you will need to copy your local files to the remote machine using SCP

eg:
    
scp -i "~/Downloads/YOUR_KEY_FILE.pem" LOCAL_FILE ubuntu@YOUR_EC2_HOST:.

#### Broken down:
scp: server copy  
-i "~/Downloads/YOUR_KEY_FILE.pem": identify yourself using your publickey  
LOCAL_FILE: the file you want to copy from your local machine  
ubuntu@YOUR_EC2_HOST: (including the :): the remote machine you want to copy to  
.: don't rename the file on the remote machine.  

### Note for the flask app file

Make sure you have the following code at the bottom of your flask app:

```python
if __name__ == '__main__':
    app.run(host='0.0.0.0', port=80)
```

### Files to copy

1. Your flask app .py file
2. the model saved as a .pkl file


### Running the app on your ec2 server

The app will be in your home directory, from there you will need to run:

`sudo python3 app.py`

this should then show you that the app is running

you can then connect to it using the address to your ec2 instance listed in the ec2 dashboard
