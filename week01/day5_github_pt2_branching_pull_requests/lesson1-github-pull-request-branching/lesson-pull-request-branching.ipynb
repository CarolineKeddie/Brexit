### What is a pull request ?

- After adding, committing, and pushing, if youre collaborating on a project, one of the common way to **share your code** is to do a pull request (PR) 
- At that point, whoever youre working with **will look over the code youve issued a Pull Request on**, and make any comments/suggestions they have, or ask any questions
    - If hosted through Github (or some similar system), these typically **take place through the browser**. In it you can have a discussion about pieces of the code and even view the changes made (see the pictures in the following slides)
- Once your code is ready to merge, the other person (or you if you have the correct rights) will merge via the browser (ie: accept the Pull Request, and then close it)

# The ugly duckling: Merge Conflict:


Sometimes you get merge conflicts when merging or pulling from a branch. Git will then tell you something like

```
CONFLICT (content): Merge conflict in Fakefile
```

It also tells you to fix the conflicts and then to commit the result. **So how to fix the conflicts?**

# What is a merge conflict? and how does it happen? 

A merge conflict usually occurs when your current branch and the branch you want to merge into the current branch **have diverged**. That is, you have commits in your current branch which **are not in the other branch, and vice versa**.

When Git merges the other branch into your current branch, it looks at the differences between latest common commit. When there are unambiguous differences (i.e. only one side changed a certain piece of code), **the changes are applied**.

The merge conflicts occur when there are **disagreeing changes** (typically, both branches have new commits). In that case, your conflicted file will have so-called conflict markers:
```
<<<<<<< HEAD
jars/VIB-lib.jar
=======
plugins/Aymeric_Plugin.jar
>>>>>>> branch-with-amazing-new-feature
```

# Resolving merge conflicts

### First solution: tell git which file you want to keep

- for your version:
```
git checkout --ours <file_with_merge_conflict>
```

- for their version:
```
git checkout --theirs <file_with_merge_conflict>
```

### Second solution: manually edit the file 

When you have a merge conflict, git will automatically point the differences in the files using `<<<<<<` or `>>>>>>` (also called conflict markers) to indicate which changes belongs to who.

- You can consequently either:
    - change manually the contents of the file
